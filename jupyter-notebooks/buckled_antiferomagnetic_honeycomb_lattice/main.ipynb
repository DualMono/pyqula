# This example shows how to compute the band structure of a buckled honeycomb lattice with charge-density wave, antiferromagnetism and spin-orbit coupling
### Import the necessary libraries

```python
# Add the root path of the pyqula library
import os ; import sys
sys.path.append(os.getcwd()+"/../../src")


# import numpy and pyqula
import numpy as np
from pyqula import geometry
```

### Create the geometry of the buckled honeycomb lattice model, and add kinetic energy, sublattice imbalance, and spin-orbit coupling 
$$ H = H_{kin} + H_{AF} + H_{CDW} + H_{SOC} $$


```python
g = geometry.buckled_honeycomb_lattice() # create the geometry
h = g.get_hamiltonian() # create hamiltonian of the system
h.add_sublattice_imbalance(0.1) # add sublattice imbalance 
h.add_antiferromagnetism(0.6) # add antiferromagnetism
h.add_soc(0.06) # add spin-orbit coupling 
```

### Diagonalize the Bloch Hamiltonian
The following function will obtain the Bloch eigenvalues $\epsilon_k$ defined as
$$ H |\Psi_k \rangle = \epsilon_k |\Psi_k\rangle $$

```python
(k,e,sz) = h.get_bands(operator="sz") # get the bandstructure
```

### Plot the Bloch band structure in the path

```python
import matplotlib.pyplot as plt
plt.scatter(k/max(k),e,c=sz,cmap="coolwarm") # do a scatter plot
plt.ylabel("Energy") # set the y label 
plt.xticks([0.,1/3.,0.5,2./3.,1.],["$\Gamma$","$K$","$M$","$K'$","$\Gamma$"])
plt.show() # show the plot
```
